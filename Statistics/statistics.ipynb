# Statistics

- Statistics helps us turn raw numbers into useful information.
- Statistics is the practice of making sense of data so we can understand what is happening and make better decisions.
- It is not just about numbers—it is about asking the right questions, organizing data, finding patterns, and drawing conclusions.
- Example: A teacher collects students’ marks. Using statistics, the teacher can: Calculate the average score, Measure how spread out the marks are, Identify students who need extra help, Evaluate if the exam was too easy or too hard

### Why Statistics is Important

- Summarize data (mean, median, charts)
- Compare groups (this month vs last month)
- Find patterns (trends, relationships)
- Make decisions based on evidence, not guesswork

### Types of Statistics

1. Descriptive Statistics
2. Inferential Statistics

#### 1. Descriptive
- What does the data say?
- Descriptive statistics is used to summarize and describe the data you already have.
- It does not predict or generalize.
- It only explains what is in front of you.
- Mean (average), Median, Mode, Range, Standard deviation, Tables, charts, graphs
- Example: A teacher collects marks of 30 students.Using descriptive statistics, the teacher finds: Average score = 72, Highest score = 95,Lowest score = 40

#### 2. Inferential
- What can we conclude beyond the data?
- Inferential statistics is used to make conclusions or predictions about a larger group using a small sample.
- “Based on this data, what can we say about the whole population?”
- Probability, Sampling, Regression
- Example: A polling agency surveys 1,000 voters instead of all citizens. Using inferential statistics, they: Predict which party is likely to win, Estimate percentage of votes

### Introduction to basic Statistics terms


#### Descrete and continuous
- Descrete: You can count outcomes
- Uses PMF (Probability Mass Function)
- Examples: Number of heads, Number of emails, Pass/fail
- Infinite values in a range
- Uses PDF (Probability Density Function)
- Examples: Height, weight, time, temperature
